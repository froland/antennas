```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numexpr
%matplotlib inline
import seaborn as sns
sns.set()
plt.rcParams["figure.figsize"] = [8, 6]
from matplotlib.ticker import EngFormatter
```

# S11
## 800 MHz - 2.5 GHz

```python
ant01 = pd.read_csv('../data/01-1.csv', skiprows=3, usecols=[0, 1], names=["frequency", "amplitude"], dtype={'frequency': np.int64, 'amplitude': np.float32}, index_col=0)
ant02 = pd.read_csv('../data/02-1.csv', skiprows=3, usecols=[0, 1], names=["frequency", "amplitude"], dtype={'frequency': np.int64, 'amplitude': np.float32}, index_col=0)
ant03 = pd.read_csv('../data/03-1.csv', skiprows=3, usecols=[0, 1], names=["frequency", "amplitude"], dtype={'frequency': np.int64, 'amplitude': np.float32}, index_col=0)
ant04 = pd.read_csv('../data/04-1.csv', skiprows=3, usecols=[0, 1], names=["frequency", "amplitude"], dtype={'frequency': np.int64, 'amplitude': np.float32}, index_col=0)
ants = pd.DataFrame()
ants['ant01'] = ant01.amplitude
ants['ant02'] = ant02.amplitude
ants['ant03'] = ant03.amplitude
ants['ant04'] = ant04.amplitude
fig, ax = plt.subplots()
formatter = EngFormatter(unit='Hz')
ax.xaxis.set_major_formatter(formatter)
ax.plot(ants)
ax.legend(['#1', '#2', '#3', '#4'], loc='lower left')
ax.set_xlabel("frequency")
ax.set_ylabel("amplitude")
plt.tight_layout()
plt.show();
```
